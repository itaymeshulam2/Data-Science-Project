```python
url="https://www.kipa.co.il/%d7%91%d7%aa%d7%99-%d7%9b%d7%a0%d7%a1%d7%aa/"
response1 = requests.get(url)
soup = BeautifulSoup(response1.content, "html.parser")

mtag=soup.find("div",attrs={"class":"col-sm-4 col-xs-12 pull-right text-left"})

temp_Cities = []
Cities = []

temp_Cities.append(str(mtag))

temp_Cities = ''.join(temp_Cities) 
mtag = re.search(r">+<",temp_Cities)
results=re.findall("\">.+?<",temp_Cities)

Cities = ''.join(results)
Cities = Cities.replace('"','').replace('>','').split('<')
print (Cities)
```

```python
linksToPages = []
url="https://www.kipa.co.il/%D7%91%D7%AA%D7%99-%D7%9B%D7%A0%D7%A1%D7%AA/"

for i in Cities: 
    
  if ' ' in i:
    i= i.replace(' ', '-')
  linksToPages.append(url + (i))

linksToPages.remove('https://www.kipa.co.il/%D7%91%D7%AA%D7%99-%D7%9B%D7%A0%D7%A1%D7%AA/')
len (linksToPages)  
```

```python
f = open("linksToSynagogue.txt", "a")
f1 = open("Cities_without_synagogue.txt", "a")
z = -1 
Cities_without_synagogue = []
for i in linksToPages:
  linksToSynagogue = []
  z = z + 1
  print (z)

  Response = requests.get(i)
  soup = BeautifulSoup(Response.content, "html.parser")
  mtag=soup.find("div",attrs={"class":"syn_city_listing"})
  templinksToSynagogue=([t['data-href'] for t in mtag.findAll("tr",attrs={"class":"clickable-row"})])
  if  len(templinksToSynagogue) == 0:
    Cities_without_synagogue.append(Cities[z])
    tempcity  = Cities[z]
    f1.write(tempcity)
    f1.write("\n")
    continue
  j = 1
  linksToSynagogue = linksToSynagogue + templinksToSynagogue 
  while True:
    j = j + 1
    
    response1 = requests.get(i + '/page/' + str(j))
    soup = BeautifulSoup(response1.content, "html.parser")
    mtag=soup.find("div",attrs={"class":"syn_city_listing"})
    temp = []
     
    for x in mtag:
      temp.append(str(mtag))
      temp_mtag = ''.join(temp) 
      temp_mtag = re.search(r"שם בית הכנסת",temp_mtag)
    
    if temp_mtag != None:
      mtag = soup.find("div",attrs={"class":"syn_city_listing"})
      linksToSynagogue1 = [t['data-href'] for t in mtag.findAll("tr",attrs={"class":"clickable-row"})]
      linksToSynagogue = linksToSynagogue + linksToSynagogue1
   
    else:
      linksToSynagogue = ''.join(linksToSynagogue)
      linksToSynagogue = linksToSynagogue.split("https://")
      del linksToSynagogue[0]

      for templink in linksToSynagogue:
        f.write(templink)
        f.write("\n")
      
      break

f.close()
f1.close() 
```

```python
def name_function(soup1):
  name = None
  if soup1.find("h1",attrs={"itemprop":"name"}) != None:
    name = soup1.find("h1",attrs={"itemprop":"name"}).get_text()
    name = name.replace('בית​‌‌​‌​‌‌​‌‌​‌​​‌​‌‌‌​​​​​‌‌​​​​‌ כנסת','').split(',')
    name = name[0]
    return name
  return name
```

```python
def function_address(address,len_address):
  temp = []
  
  for i in range(0,len_address-1):
    temp_address = ''.join(address[i]) 
    temp =''.join(temp) 
    temp = temp + ' ' + temp_address
  return temp

def city_and_address(soup1):
    city = None
    
    if soup1.find("h1",attrs={"itemprop":"name"}) != None:
        city = soup1.find("h1",attrs={"itemprop":"name"}).get_text()
        city = re.search(r"(,)(.+)",city)
        city = city.group(2)

    
    address = soup1.find("div",attrs={"class":"address detail"}).get_text()
    address = address.replace('כתובת:','').replace('נווט לבית הכנסת','').split()
    len_address = len(address)
    
    if city == None:
        city = address[len_address-1]
    address = function_address(address,len_address)
    return city , address
```

```python
def nosach_function(soup1):
  nosach = soup1.find("div",attrs={"class":"prayer detail"}).get_text()
  nosach = nosach.replace('נוסח תפילה', '')
  return nosach
```

```python
def prayer_times1(soup1):
  
  Shacharit_prayer = None
  Mincha_prayer = None
  Aravit_prayer = None
  Shacharit_Shabat_prayer = None
  Mincha_Shabat_prayer = None
  Aravit_Shabat_prayer = None

  if soup1.find("table",attrs={"class":"table table-striped"}):
    prayer_times = soup1.find("table",attrs={"class":"table table-striped"}).get_text()
    prayer_times = prayer_times.replace(' ','')
    
    if 'שחריתבחול' in prayer_times:
      Shacharit_prayer = re.search(r"(שחריתבחול)(.....)",prayer_times)
      Shacharit_prayer = Shacharit_prayer.group(2)
    
    if 'מנחהבחול' in prayer_times:
      Mincha_prayer = re.search(r"(מנחהבחול)(.....)",prayer_times)
      Mincha_prayer = Mincha_prayer.group(2)
    
    if 'ערביתבחול' in prayer_times:
      Aravit_prayer = re.search(r"(ערביתבחול)(.....)",prayer_times)
      Aravit_prayer = Aravit_prayer.group(2)

    if 'שחריתבשבת' in prayer_times:
      Shacharit_Shabat_prayer = re.search(r"(שחריתבשבת)(.....)",prayer_times)
      Shacharit_Shabat_prayer = Shacharit_Shabat_prayer.group(2)

    if 'מנחהבשבת' in prayer_times:
      Mincha_Shabat_prayer = re.search(r"(מנחהבשבת)(.....)",prayer_times)
      Mincha_Shabat_prayer = Mincha_Shabat_prayer.group(2)

    if 'ערביתבשבת' in prayer_times:
      Aravit_Shabat_prayer = re.search(r"(ערביתבשבת)(.....)",prayer_times)
      Aravit_Shabat_prayer = Aravit_Shabat_prayer.group(2)
    
  
  return Shacharit_prayer , Mincha_prayer ,Aravit_prayer , Shacharit_Shabat_prayer , Mincha_Shabat_prayer ,Aravit_Shabat_prayer
```

```python
def detail_more(soup1):
    Bar_mitzvah = False
    women = False
    Access_disabled = False
   
    if soup1.find("div",attrs={"class":"detail more"}):
        more = soup1.find("div",attrs={"class":"detail more"}).get_text()
        
        Access_disabled = re.search(r"נכים",more) 
        if Access_disabled != None:
            if 'נכים' in Access_disabled.group():
                Access_disabled = True
        
        women = re.search(r"נשים",more)
        if women != None:
            if 'נשים' in women.group():
                women = True
        
        Bar_mitzvah =  re.search(r"בר מצווה",more)
        if Bar_mitzvah != None:
            if 'בר מצווה' in Bar_mitzvah.group():
                Bar_mitzvah = True
            
    return Access_disabled,women,Bar_mitzvah
```

```python
#create a Pandas Dataframe with a row for each repository and a column for each of these datums.
j = 0
flag = []
df = pd.DataFrame(columns = ['Name', 'City', 'Address' , 'Nosach', 'shacharit_prayer' , 'mincha_prayer' , 'Aravit_prayer' , 'Shacharit_Shabat_prayer' , 'Mincha_Shabat_prayer' , 'Aravit_Shabat_prayer', 'Access_disabled' , 'women' , 'Bar_mitzvah'])

for i in range(len(content)):
  print (i) 
  x = requests.get("https://" + content[i])
  x = BeautifulSoup(x.content , "html.parser")
  name = name_function(x)
  if name == None:
    flag.append(i)
    continue
  if 'מקווה' in name:
    continue
    
  city , address = city_and_address(x)
  nosach = nosach_function(x)
  shacharit_prayer , mincha_prayer , aravit_prayer ,shacharit_Shabat_prayer ,mincha_Shabat_prayer , aravit_Shabat_prayer = prayer_times1(x)
  Access_disabled,women,Bar_mitzvah = detail_more(x)
    
  df.loc[j] = [name, city, address , nosach , shacharit_prayer, mincha_prayer , aravit_prayer ,shacharit_Shabat_prayer ,mincha_Shabat_prayer , aravit_Shabat_prayer, Access_disabled ,women , Bar_mitzvah]
  j = j + 1

f.close()
df.to_csv('project_info.csv')
```

```python
word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='word',
    token_pattern=r'\w{1,}',
    ngram_range=(1,1),
    dtype=np.float32,
    max_features=500
)

# Character Stemmer
char_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    strip_accents='unicode',
    analyzer='char',
    ngram_range=(1, 4),
    dtype=np.float32,
    max_features=100
)
```

```python
word_vectorizer.fit(df_cll.Name)
char_vectorizer.fit(df_cll.Name)
```

```python
# Train
train_word_features_name = word_vectorizer.transform(df_cll.Name)
train_char_features_name = char_vectorizer.transform(df_cll.Name)

#Sparse Matrix
train_features = hstack([train_char_features_name,train_word_features_name])

model_NB = MultinomialNB()

X_train, X_test, y_train, y_test = train_test_split(train_features, df_cll.Nosach, train_size=0.75)


model_NB = MultinomialNB()
X_train, X_test, y_train, y_test = train_test_split(train_features, df_cll.Nosach, train_size=0.75)
model_NB.fit(X_train, y_train)
predictions_tfidf = model_NB.predict(X_test)
accuracy = accuracy_score(y_test, predictions_tfidf)
accuracy
```
